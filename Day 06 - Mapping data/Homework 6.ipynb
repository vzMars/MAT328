## Question 1
Create a map of New York showing the locations of all the CUNY colleges.  Each marker should be labeled with the name of the college. 

The coordinates for the CUNY colleges are on NYC Open Data: [https://data.ny.gov/Education/City-University-of-New-York-CUNY-University-Campus/irqs-74ez/data](https://data.ny.gov/Education/City-University-of-New-York-CUNY-University-Campus/irqs-74ez/data)

URL for CSV file:  [https://raw.githubusercontent.com/megan-owen/MAT328-Techniques_in_Data_Science/main/data/CUNY_Campus_Locations.csv](https://raw.githubusercontent.com/megan-owen/MAT328-Techniques_in_Data_Science/main/data/CUNY_Campus_Locations.csv)


- ### Imports and reading csv

```python
import folium
import pandas as pd
%matplotlib inline
```

```python
colleges = pd.read_csv("https://raw.githubusercontent.com/megan-owen/MAT328-Techniques_in_Data_Science/main/data/CUNY_Campus_Locations.csv")
```

```python
colleges
```

- ### Checking for missing data

```python
colleges.describe()
```

```python
colleges.describe(include = ["O"])
```

```python
colleges = colleges.dropna()
colleges = colleges.reset_index(drop = True)
colleges
```

- ### Map of CUNY colleges

```python
colleges_map = folium.Map([40.7128, -74.0060])
for i in range(len(colleges)):
    row = colleges.loc[i]
    folium.Marker([row["Latitude"], row["Longitude"]],
                  tooltip = row["Campus"]
                 ).add_to(colleges_map)
colleges_map
```

## Question 2
As in Question 1, create a map of New York showing the locations of all CUNY College, labeled by the name of the college.

For this map, the senior colleges should be one color and all other colleges should be a different color (your choice of colors).

<details><summary>Hint:</summary>
There are multiple ways to do this question.  One way is to use filters to make smaller datasets and plot each one separately.  Another way is using conditional (if) statements.
</details>

```python
colleges_map2 = folium.Map([40.7128, -74.0060])
for i in range(len(colleges)):
    row = colleges.loc[i]
    if(row["College or Institution Type"] == 'Senior Colleges'):
        iconColor = 'green'
    else:
        iconColor = 'red'
    folium.Marker([row["Latitude"], row["Longitude"]],
                  tooltip = row["Campus"],
                  icon=folium.Icon(color=iconColor)
                 ).add_to(colleges_map2)
colleges_map2
```
